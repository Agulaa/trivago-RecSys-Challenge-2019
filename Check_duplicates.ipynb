```python
def drop_duplicates(data):
    print(data)
    user = data.iloc[0]['user_id']
    return data[data.user_id == user]
```

```python
session_user_group_2.apply(drop_duplicates)
```

```python
new_df = session_user_group_2.apply(drop_duplicates)
```
